```python
import pandas as pd
import re 
```

```python
df=pd.read_csv("data.csv")
documents = []
for i, row in df.iterrows():
    documents.append({
        "id": i,
        "title": row["title"],
        "text": row["content"]
    })
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)        # remove HTML
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text
def clean_text_transformer(text):
    return text.strip()
```

```python
unique_docs = list({doc["text"]: doc for doc in documents}.values())
```

```python
def chunk_text(text, chunk_size=200):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size):
        chunks.append(" ".join(words[i:i+chunk_size]))
    return chunks
```
